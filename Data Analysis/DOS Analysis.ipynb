```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df=pd.read_csv(r"..\CIC-IDS2017-Dataset\DoS-Wednesday-no-metadata.parquet.csv")
```

```python
# Attacks Label
plt.figure(figsize=(10,6))
sns.countplot(
    data=df,
    x='Label',
    order=df['Label'].value_counts().index,
    palette='Set2'
)
plt.title("Distribution of Network Traffic Labels",fontsize=16)
plt.xlabel("Traffic Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Protocol Distribution
protocol_map={
    6:"TCP",
    17:'UDP',
    1:'ICMP'
}
df['Protocol_Name']=df['Protocol'].map(protocol_map)
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x='Protocol_Name',
    palette='viridis'
)
plt.title("Protocol Distribution")
plt.show()
```

```python
# Distribution of important numerical features
features=['Flow Duration','Flow Packets/s','Flow Bytes/s','Packet Length Mean']
for feature in features:
    plt.figure(figsize=(8,4))
    sns.histplot(df[feature],bins=40,kde=True)
    plt.title(feature)
    plt.show()
```

```python
# Flow packets/s Analysis(Boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(
    data=df,
    x='Label',
    y='Flow Packets/s',
    palette='Set2'
)
plt.yscale('log')
plt.title("Flow Packets/s by Traffic Type",fontsize=16)
plt.xlabel("Traffic Label")
plt.ylabel("Flow Packets/s(Log Scale)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot(
    data=df,
    x='Label',
    y='Flow Packets/s',
    palette='viridis',
    cut=0
)
plt.yscale('log')
plt.title("Distribution of Flow Packets/s")
plt.xticks(rotation=45)
plt.show()
```

```python
#Flow Bytes per Second Analysis(Boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(
    data=df,
    x="Label",
    y="Flow Bytes/s",
    palette='Set3'
)
plt.yscale('log')
plt.title("Flow Bytes/s by Traffic Types")
plt.xticks(rotation=45)
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot(
    data=df,
    x='Label',
    y='Flow Bytes/s',
    palette='coolwarm',
    cut=0
)
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Distribution of Flow Bytes/s")
plt.show()
```

```python
#Forward Inter Arrival Time Analysis(Boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(
    data=df,
    x='Label',
    y='Fwd IAT Min',
    palette='Spectral'
)
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Minimum Forward Inter Arrival Time")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot(
    data=df,
    x='Label',
    y='Fwd IAT Min',
    palette='magma',
    cut=0
)
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Distribution of Fwd IAT Min")
plt.show()
```

```python
#Flow Duration Analysis(Boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(
    data=df,
    x='Label',
    y='Flow Duration',
    palette='Pastel1'
)
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Flow Duration By Traffic Type")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot(
    data=df,
    x='Label',
    y='Flow Duration',
    palette='cubehelix',
    cut=0
)
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Distribution of Flow Duration")
plt.show()
```

```python
#PAcket Structure Analysis(Boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(
    data=df,
    x='Label',
    y='Packet Length Std',
    palette='Set1'
)
plt.xticks(rotation=45)
plt.title("Packet Length Standard Deviation")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot(
    data=df,
    x="Label",
    y="Packet Length Std",
    palette='Accent',
    cut=0
)
plt.xticks(rotation=45)
plt.title("Distribution of Packet Length Variability")
plt.show()
```
